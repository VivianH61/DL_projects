```python
import cv2
import sys
import os.path
import numpy as np
from PIL import Image
count = 0
def detect(filepath, cascade_file = "/content/drive/MyDrive/Colab Notebooks/lbpcascade_animeface.xml"):
    global count
    from os.path import isfile, join
    if not os.path.isfile(cascade_file):
        raise RuntimeError("%s: not found" % cascade_file)
    # print(1)
    try:
      cascade = cv2.CascadeClassifier(cascade_file)
      image = cv2.imread(filepath, cv2.IMREAD_COLOR)
      gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
      gray = cv2.equalizeHist(gray)
      faces = cascade.detectMultiScale(gray,
                                        # detector options
                                        scaleFactor = 1.1,
                                        minNeighbors = 5,
                                        minSize = (24, 24))
      # print(2)
      #cv2.imshow("AnimeFaceDetect", image)
      for i, (x, y, w, h) in enumerate(faces):
          cropped = image[y:y+h, x:x+w]
          cv2.imwrite("/content/drive/MyDrive/Colab Notebooks/QMNX_crop/" + str(count) + ".jpg", cropped)
          count += 1

      #cv2.imshow("AnimeFaceDetect", image)
      cv2.waitKey(0)
    except:
      print("invalid input img")

    
def cropPath(filepath):
    from os import listdir
    from os.path import isfile, join
    for images in os.listdir(filepath):
      if (images.endswith(".jpg")):
          detect("./content/drive/MyDrive/Colab Notebooks/QMNX/" + images)
    # files = [f for f in listdir(filepath) if isfile(join(filepath, f))]
    # for file in files:
    #     detect(filepath, file)

filepath = "/content/drive/MyDrive/Colab Notebooks/QMNX/"
for images in os.listdir(filepath):
    if (images.endswith(".jpg")):
        detect(filepath + images)

```
